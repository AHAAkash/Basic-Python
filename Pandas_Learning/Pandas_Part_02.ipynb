# head(n) , tail(n)

if we don't provide any n it provide by default n=5

**The head() and tail() methods ** are very useful for quickly inspecting the beginning and end of a pandas DataFrame.

**df.head(n):** This method returns the first n rows of the DataFrame. If you don't specify n, it defaults to returning the first 5 rows.

**df.tail(n):** This method returns the last n rows of the DataFrame. Similar to head(), if you don't specify n, it defaults to returning the last 5 rows.

```python
print("Head of the DataFrame:")
display(df.head())

print("\nTail of the DataFrame:")
display(df.tail())
```
